```python
import pandas
from datetime import datetime
from pytz import utc
import matplotlib.pyplot as plt

data = pandas.read_csv("./data/reviews.csv", parse_dates = ['Timestamp']) # Assume pathing is not an issue
data.head()
```

## Rating Average By Day and Count

```python
data['Day'] = data['Timestamp'].dt.date
day_average = data.groupby(['Day']).mean(numeric_only=True)
day_average.head()
```

```python
day_count_by_day = data.groupby(['Day']).count()
day_count_by_day.head()
```

```python
plt.figure(figsize=(25, 3))
plt.plot(day_average.index, day_average['Rating'])
```

```python
plt.figure(figsize=(25, 3))
plt.plot(day_count_by_day.index, day_count_by_day['Rating'])
```

## Rating Average By Week

```python
data['Week'] = data['Timestamp'].dt.strftime('%Y-%U')
data
```

```python
week_average = data.groupby(['Week']).mean(numeric_only=True)
week_average.head()
# data['Week'].max()
```

```python
plt.figure(figsize=(25, 3))
plt.plot(week_average.index, week_average['Rating'])
```

## Rating Average By Month

```python
data['Month'] = data['Timestamp'].dt.strftime('%Y-%m')
data
```

```python
month_average = data.groupby(['Month']).mean(numeric_only=True)
month_average.head()
```

```python
# plt.figure(figsize=(25, 3))
plt.plot(month_average.index, month_average['Rating'])
```

## Rating Average By Year

```python
data['Year'] = data['Timestamp'].dt.strftime('%Y')
year_average = data.groupby(['Year']).mean(numeric_only=True)
plt.figure(figsize=(25, 3))
plt.plot(year_average.index, year_average['Rating'])
```
